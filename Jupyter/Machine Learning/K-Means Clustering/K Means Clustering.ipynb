```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=200,n_features=2, centers=4, 
                cluster_std=1.8, random_state=101)
data
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1], c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans= KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

In a real case we would be done for now and could not do the next step

```python
# Two plots, the first based on the model's calculations, the second one based on the original data
fig, (ax1,ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c= kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c= data[1], cmap='rainbow')
```

```python
kmeans2= KMeans(n_clusters=2)
kmeans2.fit(data[0])

# Two plots, the first based on the model's calculations, the second one based on the original data
fig, (ax1,ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c= kmeans2.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c= data[1], cmap='rainbow')
```
