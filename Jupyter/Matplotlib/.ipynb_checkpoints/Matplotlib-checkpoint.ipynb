```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y= x**2
```

# Functional Programming

```python
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
# plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object oriented

```python
fig=plt.figure() # This turns the plot into a canvas where you can add stuff
axes = fig.add_axes([0,0,0.8,0.8]) 
axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,0.3,0.4]) # Easier to see the changes. The first coordinates are the x0 and y0 and the last two coordinates are the height and the width

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
# axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout() # Avoids overlapping axes
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[1].plot(x,y)
```

# Figure Size and DPI

```python
fig= plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(x,y)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi=200)
fig.savefig('my_picture.jpeg')
```

```python
fig= plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared') 
ax.plot(x,x**3, label='X Cubed')

ax.set_title('Title')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')

ax.legend(loc=0) # The legend only works if there is a label in the ax.plot()
# The loc in the legend puts the legend wherever you want 
```

# Plot Appearance

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])

# ax.plot(x,y,color='orange')
ax.plot(x,y,color='#FF8C00') # RGB Hex Code
```

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', linewidth=5, alpha=0.8, linestyle='--', marker='o',
        markersize=20, markerfacecolor='yellow', markeredgewidth=5, markeredgecolor='green')
```

# Various line and marker styles

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', lw=2, ls='--')

ax.set_xlim=([0,1])
ax.set_ylim=([0,2])
```
